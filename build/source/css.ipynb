## position
```css
定位
position:static;静态定位（默认定位方式），一般用来清除定位
position：relative;相对定位(自恋型），相对于自己标准流的位置,原来的位置继续保留，不脱标
position：absolute;绝对定位，完全脱标的：a）父级无定位，b)父级无定位; 子绝父相
position: fixed;固定定位，相对于浏览器，完全脱标，不占位置
叠放次序：只有相对定位、绝对定位、固定定位才有z-index,取值越大显示靠上
定位模式转换：元素添加绝对定位、固定定位后会转换为inline-block
元素显示与隐藏：a)display:none; 
/*元素隐藏，隐藏后不占位置,显示使用block*/
visibility: hidden; /*元素隐藏，隐藏后占位置,显示使用visible*/


```

    1. vscode安装easy less插件,可实现less文件自动编译为css文件（纯编写less文件，想查看转换后的css可使用)；禁止less自动编译方法：
    ①. vscode设置easyLess： File -> Preferences -> settings -> 搜索设置输入‘easy less’-> 点击Edit in setting.json，新增以下内容
        "less.compile": {
            "compress": false, // true => remove surplus whitespace
            "sourceMap": false, // true => generate source maps (.css.map files)
            "out": false, // false => DON'T output .css files (overridable per-file, see below)
        }
    ②. out 配置编译路径设为false可以禁止自动编译less文件，当需要编译某个文件时，只需要在文件最上放添加 "//out 输出的css文件路径":



    2. html中直接引入less文件，使用less.min.js文件

三. Less视频笔记
